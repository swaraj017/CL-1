```python
avg_price_by_age = df.groupby("House_Age")["House_Price_Unit_Area"].mean()
avg_price_by_age.head()
```

```python
Q1 = df["House_Price_Unit_Area"].quantile(0.25)
Q3 = df["House_Price_Unit_Area"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["House_Price_Unit_Area"] < lower_bound) | (df["House_Price_Unit_Area"] > upper_bound)]
outliers
```

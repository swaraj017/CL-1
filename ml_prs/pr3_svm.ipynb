```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
```

```python
digits = datasets.load_digits()
```

```python
print("Dataset shape:", digits.data.shape)
print("Number of labels:", len(digits.target))
```

```python
plt.figure(figsize=(8, 4))
for i in range(8):
    plt.subplot(2, 4, i + 1)
    plt.imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f"Label: {digits.target[i]}")
    plt.axis('off')
plt.show()
```

```python
# Step 3: Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.2, shuffle=True, random_state=42
)
```

```python
# Step 4: Create SVM classifier
classifier = svm.SVC(gamma=0.001)   # small gamma for smooth decision boundary

# Step 5: Train the classifier
classifier.fit(X_train, y_train)

# Step 6: Predict labels for test set
y_pred = classifier.predict(X_test)
```

```python
# Step 8: Plot Confusion Matrix (Easy Way)
cm = metrics.confusion_matrix(y_test, y_pred)
plt.imshow(cm, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
